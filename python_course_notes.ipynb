### List comprenhshion
unique most featue , here are some examples 

```python
sq = [n**2 for n in range(10)]
print(sq)
# combined condition in the list only otherwise we would have done something like this
squares = []
for n in range(10):
    squares.append(n**2)
print(squares)
```

```python
# conditions can also be inserted too in that 
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
short_planets = [planet for planet in planets if (len(planet) < 6)]
short_planets
```

```python
# .upper() converts all the charachter of the string into uppercase 
loud_short_planets = [
                        planet.upper() + '!' 
                        for planet in planets 
                        if len(planet) < 6
                        ]
print(loud_short_planets)
#otherwise the thing in the bracket would look like [planet.upper() + '!' for planet in planets if len(planet) < 6]
[32 for planet in planets]
# value_is_this for iterator in list 
# this is the analogy where planet is used as the iterator to point the value on the list but there is 32 instead of the planet that is why 32x5
```
